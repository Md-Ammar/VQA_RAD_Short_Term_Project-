```python
def pre_process(sentence:str):
    try:
        sen = nlp(sentence)
        tokens = [token.text.lower() for token in sen if not token.is_punct]
        re_sen = " ".join(tokens)
        re_sen.replace('?' , " ")
        return re_sen
    except:
        print(sentence)
        return str(sentence)
```

```python
df['question'] = df['question'].apply(pre_process)
```

# Fine Tuning VILT VQA 

```python
id2label = {i: label for i, label in enumerate(le.classes_)}
label2id = {label: i for i, label in enumerate(le.classes_)}
```

```python
label2id
```

```python
from transformers import AutoProcessor, ViltForQuestionAnswering

processor = AutoProcessor.from_pretrained("dandelin/vilt-b32-finetuned-vqa")
model = ViltForQuestionAnswering.from_pretrained("dandelin/vilt-b32-finetuned-vqa" ,label2id=label2id ,id2label = id2label ,num_labels=516,ignore_mismatched_sizes=True)
```

```python
df["answer"].nunique()
```

```python
import torch
from PIL import Image
def pre_process(data):
    image = Image.open(data["image_name"]).convert("RGB")
    image = image.resize((448, 448))
    encoding = processor(image, data["question"],padding="max_length", 
        truncation=True,return_tensors="pt")
    for k, v in encoding.items():
          encoding[k] = v.squeeze()

    num_labels = len(le.classes_)  
    target = torch.zeros(num_labels)
    target[data['answer']] = 1.0  
    
    encoding["labels"] = target
    
    return encoding
```

```python
dataset = full_data.map(pre_process)
```

```python
dataset
```

```python
dataset = dataset.remove_columns(['image_name', 'question', 'answer', 'token_type_ids', 'pixel_mask'])
dataset
```

```python
dataset
```

```python
from transformers import  Trainer, TrainingArguments 
training_args = TrainingArguments(
    output_dir="./vilt-vqa-rad-finetuned",
    evaluation_strategy="steps",
    learning_rate=5e-5,
    per_device_train_batch_size=4,
    num_train_epochs=3,
    save_steps=500,
    save_total_limit=2,
)
```

```python
# Define Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    eval_dataset=dataset 
)

# Train the model
trainer.train()
```

```python
trainer.save_model("./vilt-vqa-rad-finetuned-AIC")  # You can specify the path where you want to save the model
processor.save_pretrained("./vilt-vqa-rad-finetuned-AIC")
```
